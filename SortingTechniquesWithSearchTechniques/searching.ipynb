# Linear Search
- Best case O(1)
- Worst Case O(n)

```python
l1 = [10,2, 5, 8, 3, 4]

key = int(input('Enter the target : '))
found  = False

for i in range(len(l1)):

    if key== l1[i]:

        print(f' The Key {key} is found at index {i}')
        found = True
        break

if found == False:
    print(f'Key { key} is not found in the list')
```



### When it is asked about the 2nd occurance of the key..

### take the number of occurance to be searched..every time we find the key ,
### increase the count by 1 till we get the nth occurance..


```python
l1 = [10,2, 5, 8, 3, 5,4]

key = int(input('Enter the target : '))


nth = int(input('Enter the occurance..'))

count = 0 
found  = False

for i in range(len(l1)):

    if key== l1[i]:
        count += 1
        if count == nth:
            print(f' The Key {key} is found at index {i}')
            found = True
            break

if found == False:
    print(f'{count+1}th occurance of the key { key} is not found in the list')
```

## BINARY SEARCH

### Only works when the list is Sorted..
- Worst cAse O(logn base 2)
- Best Case O(1)

```python
l1 = [10, 20, 30 , 40 , 50 ,60]

key = int(input('Enter the target: '))

first = 0
last = len(l1) - 1

found = False

while first <= last:

    mid = (first + last ) // 2

    if l1[mid] == key:

        print(f'Key {key} is found at index {mid}')
        found = True
        break
    elif key > l1[mid]:
        first = mid+1

    else:
        last = mid - 1

if found == False:
    print(f' key {key} is not found ')
```
